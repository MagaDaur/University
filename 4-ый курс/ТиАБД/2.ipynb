```python
import pandas as pd
import plotly.graph_objs as go
from plotly.subplots import make_subplots

# Загрузка встроенного датасета из pandas (в качестве примера возьмём dataset 'iris')
df = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv')

# Вывод информации о данных
print(df.info())
print(df.head())

# Проверка на пропущенные значения и их удаление
if df.isnull().values.any():
    df = df.dropna()

# Настройка параметров для построения столбчатой диаграммы
fig = make_subplots(rows=1, cols=1)
data = go.Bar(
    x=df['species'],  # по оси X будет категориальная переменная species
    y=df['sepal_length'],  # по оси Y числовой показатель sepal_length
    marker=dict(
        color=df['sepal_width'],  # Цвет в зависимости от другого признака
        coloraxis="coloraxis"  # Настройка цветовой оси
    ),
    marker_line=dict(color='black', width=2)  # Черные границы с толщиной линии 2
)

# Добавление данных в график
fig.add_trace(data)

# Настройка заголовка
fig.update_layout(
    title={'text': 'Iris Dataset Bar Chart', 'x': 0.5, 'y': 0.95, 'xanchor': 'center', 'yanchor': 'top', 'font': {'size': 20}},
    xaxis_title="Species",
    yaxis_title="Sepal Length",
    font=dict(size=16),  # Подписи для осей
    xaxis=dict(tickangle=315),  # Разворот меток оси X под углом 315
    xaxis_tickfont=dict(size=14),
    yaxis_tickfont=dict(size=14),
    height=700,  # Высота графика
    coloraxis=dict(colorscale='Viridis'),  # Цветовая ось
    margin=dict(l=0, r=0, t=100, b=0),  # Убираем лишние отступы
    showlegend=False  # Отключение легенды
)

# Добавление сетки с ivory цветом и толщиной линии 2
fig.update_xaxes(showgrid=True, gridwidth=2, gridcolor='ivory')
fig.update_yaxes(showgrid=True, gridwidth=2, gridcolor='ivory')

# Отображение графика
fig.show()
```

```python
from plotly.subplots import make_subplots

# Группировка данных по видам растений (species)
species_count = df['species'].value_counts()

# Построение круговой диаграммы
fig = make_subplots(rows=1, cols=1)

pie = go.Pie(
    labels=species_count.index,  # Метки (вида растений)
    values=species_count.values,  # Количество экземпляров каждого вида
    marker=dict(
        line=dict(color='black', width=2)  # Черная линия с толщиной 2 для каждой доли
    ),
    textinfo='label+percent',  # Показывать метки и проценты
    textposition='inside',  # Расположение текста внутри сегментов
)

# Добавление круговой диаграммы в график
fig.add_trace(pie)

# Настройка оформления
fig.update_layout(
    title={'text': 'Iris Species Distribution', 'x': 0.5, 'y': 0.95, 'xanchor': 'center', 'yanchor': 'top', 'font': {'size': 20}},
    showlegend=False,  # Отключение легенды
    margin=dict(l=0, r=0, t=100, b=0)  # Убираем лишние отступы
)

# Отображение круговой диаграммы
fig.show()
```

```python
import matplotlib.pyplot as plt

x = df['sepal_length']
y1 = df['sepal_width']
y2 = df['petal_length']
y3 = df['petal_width']

# Построение линейных графиков
plt.figure(figsize=(10, 6))

# Линия и маркеры для sepal_width
plt.plot(x, y1, label='Sepal Width', color='crimson', marker='o', markerfacecolor='white', 
         markeredgecolor='black', markeredgewidth=2, linewidth=2)

# Линия и маркеры для petal_length
plt.plot(x, y2, label='Petal Length', color='crimson', marker='s', markerfacecolor='white', 
         markeredgecolor='black', markeredgewidth=2, linewidth=2)

# Линия и маркеры для petal_width
plt.plot(x, y3, label='Petal Width', color='crimson', marker='^', markerfacecolor='white', 
         markeredgecolor='black', markeredgewidth=2, linewidth=2)

# Настройка сетки
plt.grid(True, linewidth=2, color='mistyrose')

# Настройка легенды
plt.legend()

# Подписи осей
plt.xlabel('Sepal Length')
plt.ylabel('Measurements')

# Заголовок графика
plt.title('Dependence of Sepal and Petal Measurements on Sepal Length')

# Отображение графика
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
from tensorflow.keras.datasets import fashion_mnist

# Загрузка датасета Fashion MNIST
(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()

# Преобразуем данные в двумерные массивы (для t-SNE вход должен быть 2D)
x_train = x_train.reshape(x_train.shape[0], -1)  # Преобразуем изображения в вектор
x_train = x_train.astype('float32') / 255.0  # Нормализуем данные

# Функция для построения t-SNE с разной перплексией
def plot_tsne(perplexity):
    tsne = TSNE(n_components=2, perplexity=perplexity, n_iter=1000, random_state=42)
    x_tsne = tsne.fit_transform(x_train[:3000])  # Ограничиваем количество примеров для скорости

    plt.figure(figsize=(8, 6))
    scatter = plt.scatter(x_tsne[:, 0], x_tsne[:, 1], c=y_train[:3000], cmap='tab10', s=10)
    plt.colorbar(scatter)
    plt.title(f't-SNE с перплексией = {perplexity}')
    plt.show()

# Визуализация t-SNE для различных значений перплексии
for perplexity in [5, 30, 50]:
    plot_tsne(perplexity)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import umap
from sklearn.manifold import TSNE
import time
from sklearn.datasets import load_iris

# Загрузим данные (в данном случае используем набор данных Iris)
data = load_iris()
X = data.data
y = data.target

# Функция для визуализации результатов
def plot_results(X_embedded, title):
    plt.figure(figsize=(8, 6))
    scatter = plt.scatter(X_embedded[:, 0], X_embedded[:, 1], c=y, cmap='Spectral', s=50)
    plt.title(title)
    plt.colorbar(scatter)
    plt.show()

# Параметры для UMAP
n_neighbors_list = [5, 15, 30]
min_dist_list = [0.0, 0.1, 0.5]

# Визуализация UMAP с различными параметрами
for n_neighbors in n_neighbors_list:
    for min_dist in min_dist_list:
        start_time = time.time()
        reducer = umap.UMAP(n_neighbors=n_neighbors, min_dist=min_dist)
        X_umap = reducer.fit_transform(X)
        elapsed_time = time.time() - start_time
        
        plot_results(X_umap, f'UMAP: n_neighbors={n_neighbors}, min_dist={min_dist} (Time: {elapsed_time:.4f}s)')

# Время работы t-SNE
start_time_tsne = time.time()
tsne = TSNE(n_components=2, random_state=42)
X_tsne = tsne.fit_transform(X)
elapsed_time_tsne = time.time() - start_time_tsne

plot_results(X_tsne, f't-SNE (Time: {elapsed_time_tsne:.4f}s)')
```
