```python
import math

class TrigFig(object):
    def __init__(self, value) : self.value = float(value)
    def  cos(self): 
        try   : return math.cos(self.value)
        except Exception as e: return e
    def  sin(self): 
        try   : return math.sin(self.value)
        except Exception as e: return e
    def  tan(self): 
        try   : return math.tan(self.value)
        except Exception as e: return e
    def arcsin(self): 
        try   : return math.asin(self.value)
        except Exception as e: return e
    def arccos(self): 
        try   : return math.acos(self.value)
        except Exception as e: return e
    def arctan(self): 
        try   : return math.atan(self.value)
        except Exception as e: return e
    def deg(self): 
        try   : return math.degrees(self.value)
        except Exception as e: return e
        
number = TrigFig(1)
print("COS:", number.cos(), 
      "SIN:", number.sin(), 
      "TAN:", number.tan(), 
      "ASIN:", number.arcsin(), 
      "ACOS:", number.arccos(), 
      "ATAN:", number.arctan(), 
      "DEG:", number.deg(), sep='\n')
```

```python
class Node(object):
    def __init__(self, value, left=None, right=None):
        self.value = value
        self.left = left
        self.right = right
    
    def printTree(node, level=0):
        if node.left != None:
            node.left.printTree(level + 1)
        
        print(' ' * 4 * level + '-> ' + node.value)

        if node.right != None:
            node.right.printTree(level + 1)

t = Node("a", Node("b",Node("d"), Node("e")),Node("c", Node("f")))
t.printTree()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn.metrics import classification_report, confusion_matrix

x = np.array([[-1, -1], [-2, -1], [-3, -2], [1, 1], [2, 1], [3, 2]])
target = [0, 0, 0, 1, 1, 1]

x_train, x_test, y_train, y_test = train_test_split(x, target, test_size=0.2)

classifier = DecisionTreeClassifier()
classifier.fit(x_train, y_train)

tree.plot_tree(classifier)

y_pred = classifier.predict(x_test)

print(confusion_matrix(y_test, y_pred), classification_report(y_test, y_pred), sep='\n')
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn import metrics

url = 'https://raw.githubusercontent.com/aniruddhachoudhury/Red-Wine-Quality/master/winequality-red.csv'
dataset = pd.read_csv(url)

x_train, x_test, y_train, y_test = train_test_split(dataset.iloc[:, :-1], dataset.iloc[:, -1], test_size=0.2, random_state=0)

regressor = DecisionTreeRegressor()
regressor.fit(x_train, y_train)

tree.plot_tree(regressor)

y_pred = regressor.predict(x_test)

df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
print(df)

print(f'Mean Squared Error: {metrics.mean_squared_error(y_test, y_pred)}')
print(f'Mean Absolute Error: {metrics.mean_absolute_error(y_test, y_pred)}')
print(f'Mean Average Error: {metrics.mean_absolute_error(y_test, y_pred) / np.average(dataset.iloc[:, -1]) * 100:.1f}%')
```
