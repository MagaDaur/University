```python
import seaborn as sns; sns.set()
import warnings
import numpy as np
warnings.filterwarnings('ignore')
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

X = np.array([[5,3],
              [10,15],
              [15,12],
              [24,10],
              [30,45],
              [85,70],
              [71,80],
              [60,78],
              [55,52],
              [80,91],])

for n in range(1, 11):
    kmeans = KMeans(n_clusters = n)
    kmeans.fit(X)
    y_kmeans = kmeans.predict(X)

    print("n_clusters = {0:2}".format(n), y_kmeans)
    
    plt.scatter(X[:, 0], X[:, 1], s=20, c=y_kmeans, cmap='rainbow')
    plt.show()
```

```python
import seaborn as sns; sns.set()
import pandas as pd
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as shc
from sklearn.cluster import AgglomerativeClustering

X = pd.read_csv('https://gist.githubusercontent.com/netj/8836201/raw/6f9306ad21398ea43cba4f7d537619d0e07d5ae3/iris.csv').iloc[:, :-1]

data = X.iloc[:, 0:2].values

plt.figure(figsize=(20, 15), dpi=500)
plt.title("Customer Dendograms")
dent = shc.dendrogram(shc.linkage(data, method='ward'))


cluster = AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='ward')
cluster.fit_predict(data)

plt.figure(figsize=(10, 7))
plt.scatter(data[:,0], data[:,1], c=cluster.labels_, cmap='rainbow')
```
